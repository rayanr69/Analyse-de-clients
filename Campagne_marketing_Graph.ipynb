```python
# Fonction pour afficher les pourcentages sur les barres
def perc_on_bar(z, data):
    '''
    Plot countplot avec pourcentages.
    z : str, nom de la colonne catÃ©gorielle
    data : DataFrame, dataset contenant la colonne
    '''
    total = len(data[z]) 
    plt.figure(figsize=(15, 5))
    ax = sns.countplot(data[z], palette='Paired', order=data[z].value_counts().index)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)  # Pourcentage
        x = p.get_x() + p.get_width() / 2 - 0.05  
        y = p.get_y() + p.get_height()
        
        ax.annotate(percentage, (x, y), size=12)  # pour le pourcentage
    
    plt.show() 

perc_on_bar('Marital_Status', market)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


# Ajouter une colonne 'Age' en fonction de l'annÃ©e de naissance
market['Age'] = 2024 - market['Year_Birth']

# Appliquer les filtres d'Ã¢ge et de revenu
market_filtered = market[(market['Age'] < 90) & (market['Income'] < 200000)]

# Features for clustering
features = market_filtered[['Age', 'Income', 'Kidhome', 'Teenhome']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# K-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
market_filtered['Segment'] = kmeans.fit_predict(features_scaled)

# Visualisation des segments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Income', y='Age', hue='Segment', data=market_filtered, palette='tab10')
plt.title('Customer Segmentation')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
market_filtered = market[(market['Age'] < 90) & (market['Income'] < 200000)]

def categorize_customer(row):
    if row['Kidhome'] > 0 or row['Teenhome'] > 0:
        return 'Familles avec enfants'
    elif 25 <= row['Age'] <= 40 and row['Kidhome'] == 0 and row['Teenhome'] == 0:
        return 'Jeunes professionnels'
    elif row['Age'] >= 60:
        return 'RetraitÃ©s actifs'
    else:
        return 'Autres'

market_filtered['Segment'] = market_filtered.apply(categorize_customer, axis=1)

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Income', y='Age', hue='Segment', data=market_filtered, palette='tab10')
plt.title('Segmentation des Clients par Ã‚ge, Revenu et Structure Familiale')
plt.xlabel('Revenu')
plt.ylabel('Ã‚ge')
plt.legend(title='Segment')
plt.grid(True)
```

```python
channels = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
response_by_channel = market.groupby('Segment')[channels].mean()
response_by_channel.plot(kind='bar', figsize=(12, 8))
plt.title('Response by Marketing Channel')
plt.xlabel('Segment')
plt.ylabel('Average Number of Purchases')
plt.show()
```
